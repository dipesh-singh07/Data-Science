```python
lst = [1,2,3, 'hello', 'c', [2,4]]

print(lst)

#append()
lst.append('dipesh')
print(lst)

#extend()
lst.extend([2,5,6,7])
print(lst)

#insert(index, value)
lst.insert(3,'ritesh')
print(lst)

#pop()
lst.pop()
print(lst)

#remove -> remove 1st occurance only
lst.remove(2)
print(lst)

#index()
print(lst.index(2))

#count()
print(lst.count(2))

```

```python
lst3 = [5,7,2,4,9]

#sort()
lst3.sort()
print(lst3)

#sum,min,max
print(min(lst3))
print(sum(lst3))
print(max(lst3))

#sorted() -> returns a new list
lst4 = [7,8,45,3,1,9]
print(sorted(lst4))
```
