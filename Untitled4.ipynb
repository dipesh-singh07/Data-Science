```python
import numpy as np
```

```python
a = np.random.randint(1,91,(5,5));

print("\nOriginal Matrix:")
print(a)

print("\nrow wise mean:")
print(np.mean(a,axis=1))

print("\ncolumn wise mean:")
print(np.mean(a,axis=0))

a[a%2==0] = -1;
print("\nAfter replacing even no. to -1 :")
print(a)

row_sorted = np.sort(a, axis = 1)
print("\nRow wise sorted Matrix : ")
print(row_sorted)

column_sorted = np.sort(a, axis = 0)
print("\nColumn wise sorted Matrix : ")
print(column_sorted)

print("\nRow wise Sum : ")
print(np.sum(a, axis = 1))

print("\nColumn wise Sum : ")
print(np.sum(a, axis = 0))

#count no. of negative in each column
print("\n",np.sum(a<0,axis = 0))

```

```python
stock = np.random.randint(10,100,(5,5))
demand = np.random.randint(1,120,(5,5))
print("Stock :")
print(stock,"\n")
print("Demand :")
print(demand,"\n")

shortage = stock - demand
print("Shortage : \n",shortage)

outOfStock = np.where(shortage<1,0,shortage)
print("\nOut of Stock : \n",outOfStock)

totalShortage = np.sum(np.where(shortage<1,-1*shortage,0),axis = 0)
print("\nTotal Shortage : \n",totalShortage)

zeroStock = np.sum(np.where(outOfStock == 0,1,0),axis =0)
print("\nProducts with Zero Stock : \n",zeroStock)

sortedrow = np.sort(outOfStock,axis = 1)
print("\nSort each row based on remaining stock : \n",sortedrow)
```
