```python
import numpy as np
```

```python
# import numpy as np
a = np.array([3,2,4,5])
print(a);

#all datatypes will be same, all of them will be converted into a string
b = np.array([3,4,6,'d',True])
print(b);
```

```python
np_arr = np.array([3,3,3,3,3.3,'hello',False])
print(bool(np_arr[-1]));

a = False;
print(bool(a));
```

```python
a = np.zeros((2,3,4), dtype=int)
print(type(a[0][0][0]))

b = np.ones((2,3), dtype=int)
print(b)

c = np.empty(4, dtype=int)
print(c)

d = np.full(4,5,dtype = int)
print(d)
```

```python
a = np.arange(3,10,1,dtype=int)
print(a)

b = np.linspace(1,2,5,dtype=float) #end value is included in this
print(b)

c=np.shape(b);#Returns dimensions as a tuple.
print(c)

d=np.size(b)#Returns total number of elements.
print(d)
```

```python
a = np.random.randint(2,7,2)
print(a)

b = np.random.choice([2,4,5,6],2)
print(b)
```

```python
#broadcasting
#Broadcasting allows NumPy to perform operations on arrays of different shapes
#without copying data.
a = np.array([1,2,3,4])
b = np.array([3,4,5,6])
print(a*b)
print(a+b)
print(a%b)
print(a*2)

c = [1,2,3,4]
d = [4,5,6,7]
print(c*2)
print(c+d)
```

```python
a = np.random.randint(2,15,6)
print(a)

b = a.reshape(3,2)
print(b)

c = b.transpose()
print(c)

e = np.random.randint(2,15,6);
f = e.reshape(2,3)
d = np.concatenate((c,f), axis = 1)
print(d)
```

```python
a = np.random.randint(2,15,6)
print(a)
b = np.random.randint(2,15,6)
print(b)

c = np.where(b>8,b,a)
print(c)
```

```python
a = np.random.randint(1,10,5)
print(a)
a.sort()
print(a)

print(a.max())
print(a.min())
print(a.mean())
print(a.sum())

c = np.array([[1,2],[3,4]])
d = np.array([[2,3],[4,5]])
print(np.matmul(c,d))

print(np.std(a))
```
